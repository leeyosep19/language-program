```python
#막대그래프 : Bar()

import matplotlib.pyplot as plt

figure = plt.figure()     #전체영역 (도화지)
axes = figure.add_subplot(111)     #차트영역(1행 1열 1번째)

x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y = [8, 6, 5, 4, 7, 9, 5]
axes.bar(x,y)
plt.title('weekday call')
plt.xlabel('week')
plt.ylabel('call')
plt.show()

```

```python
#산포 그래프 scatter()

import matplotlib.pyplot as plt

figure = plt.figure()     #전체영역 (도화지)
axes = figure.add_subplot(111)     #차트영역(1행 1열 1번째)

x = [1, 2, 3, 4, 5, 6]
y = [6, 4, 1, 2, 7, 5]
area = [50, 100, 150, 200, 250, 300]  # 각점의 크기
color = ['red', 'green', 'blue', 'orange', 'aqua', 'crimson']
axes.scatter(x,y,s=area,c=color)
plt.show()
```

```python
#원형 그래프 : pie()

import matplotlib.pyplot as plt

figure = plt.figure()     #전체영역 (도화지)
axes = figure.add_subplot(111)     #차트영역(1행 1열 1번째)

data = [1,2,3]
axes.pie(data)
plt.show()
```

```python
#범례와 레이블을 추가한 원형 그래프
import matplotlib.pyplot as plt

figure = plt.figure()     #전체영역 (도화지)
axes = figure.add_subplot(111)     #차트영역(1행 1열 1번째)

data = [1, 2, 3]
label = ['Good', 'Bad', 'Normal']
axes.pie(data,labels=label,autopct='%d%%')  #레이블을 포함, 정수%로 표시
plt.legend(label,loc='center left')
plt.show()
```

```python
#page 349
import random

figure = plt.figure(figsize=(20,6))
axes = figure.add_subplot(111)

x = [n for n in range(101)]    #리스트 내포 (리스트 컴프리헨션)
y1 = []    #꺽은선
y2 = []    #막대
for i in range(101):
  y1.append(random.randint(0,100))    #0~100 정수 중 하나
  y2.append(random.randint(0,100))
axes.plot(x, y1, color='r',marker='.')   #꺽은선
axes.bar(x,y2,color='g',)             #막대
plt.show()

```
